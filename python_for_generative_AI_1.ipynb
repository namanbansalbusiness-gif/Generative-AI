```python
#while
i = 0 #initialization
while i<5:
  print(i)
  i+=1
```

```python
#break--this statement break the loop
i = 0 #initialization
while i<5:
  i+=1
  if i == 3:
    break
  else:
    print(i)
```

```python
#continue statement -- skip the current iteration
i = 0
while i<5:
  i+=1
  if i == 3:
    continue
    print(i)
  else:
    print(i)
```

```python
#pass statement --- pass the code
i = 0
while i<5:
  i+=1
  if i == 3:
    pass
    print(i)
  else:
    print(i)
```

```python
#pie plot
import numpy as np
import matplotlib.pyplot as plt

# Data
fruit_percentage = np.array([20, 25, 35, 20])
fruit_name = np.array(["Apple", "Banana", "Orange", "Watermelon"])
colors = np.array(["red", "yellow", "orange", "green"])
myexplode = np.array([0.2, 0, 0, 0])  # Corrected np.array

# Plot
plt.pie(fruit_percentage, labels=fruit_name, autopct="%1.1f%%", colors=colors, explode=myexplode)
plt.legend(title="Fruit", loc="best")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
age_of_car = np.array([4, 5, 6, 7, 8, 9, 10,11,12 ])
speed_of_car = np.array([150, 140, 130, 120, 110, 100, 90, 80, 70])  # Fixed typo 'S0'  '80'

# Plot
plt.scatter(age_of_car, speed_of_car, color = 'red')
plt.xlabel("Age of Car")  # Fixed typo '*-ge ef car'
plt.ylabel("Speed of Car")  # Fixed typo 'Speed of ce'
plt.title("Age vs Speed of car")
plt.grid(True)
plt.show()
```

```python
#histogram -- It showing the frequency distribution of the
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
x = np.random.randint(40, 100, (1, 200))  # Random integers between 40 and 100
x = x.flatten()  # Convert to 1D array

# Print the data
print(x)

# Plot histogram
plt.hist(x, bins=10, color='skyblue', edgecolor='black')
plt.title("Histogram of Car Ages")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
